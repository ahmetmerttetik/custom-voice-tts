```python
import sys
from piper import PiperVoice
import time
import wave
from pathlib import Path
import logging
from concurrent.futures import ThreadPoolExecutor, as_completed

logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')

def load_voice(model_path, config_path):
    try:
        return PiperVoice.load(model_path=model_path, config_path=config_path)
    except Exception as e:
        logging.error(f"Ses modeli yüklenirken hata oluştu: {e}")
        sys.exit(1)

def synthesize_chunk(voice, text, synthesize_args, output_path):
    try:
        with wave.open(str(output_path), "wb") as wav_file:
            voice.synthesize(text, wav_file, **synthesize_args) # kwargs kullanılmış
        return output_path
    except Exception as e:
        logging.error(f"Ses sentezlenirken hata oluştu: {e}")
        return None

def main(model_path, config_path, output_dir, text, chunk_size=1000):
    t0 = time.perf_counter()

    voice = load_voice(model_path, config_path)

    synthesize_args = {
        "length_scale": 0.95,
        "noise_scale": 0.5,
        "noise_w": 0.8,
        "sentence_silence": 0.00,
    }

    output_dir = Path(output_dir)
    output_dir.mkdir(parents=True, exist_ok=True)

    # Metni parçalara ayır
    chunks = [text[i:i+chunk_size] for i in range(0, len(text), chunk_size)]

    with ThreadPoolExecutor() as executor:
        futures = []
        for i, chunk in enumerate(chunks):
            output_path = output_dir / f"chunk_{i}_{int(time.time() * 1e6)}.wav"
            futures.append(executor.submit(synthesize_chunk, voice, chunk, synthesize_args, output_path))

        # Tüm parçaların işlenmesini bekle
        completed_paths = []
        for future in as_completed(futures):
            result = future.result()
            if result:
                completed_paths.append(result)

    t1 = time.perf_counter()
    logging.info(f"Toplam sentez süresi: {t1-t0:.3f} saniye")
    logging.info(f"Oluşturulan dosyalar: {', '.join(str(path) for path in completed_paths)}")

if __name__ == "__main__":
    model = "/home/mert/Desktop/piper/model_6602.onnx"
    config = "/home/mert/Desktop/piper/model_6602.onnx.json"
    output = '.'
    text   = """
    Bu ne harika bir gün!" dedi zürafa. "Evet," diye yanıtladı tilki, "böyle günlerde hayat çok güzel."
    Ancak; bir gün, ani bir fırtına çıktı! Yağmur şiddetle yağdı, gök gürledi ve şimşek çaktı. Herkes hızlıca evlerine döndü: çünkü fırtına tehlikeliydi. Çiftçi, "Herkes dikkatli olsun!" diye bağırdı.
    O akşam, herkes kendi evinde toplanmıştı. Biraz kitap okudular, biraz sohbet ettiler, biraz da dinlendiler... Ve böylece, uzun ve maceralı bir gün sona erdi.
    """

    main(model, config, output, text)


# class yapısı oluştur.
# class -> name sesli asistan (sound assistan)
# metotlar -> synthesize_args = {
    #     "length_scale": 0.95,
    #     "noise_scale": 0.5,
    #     "noise_w": 0.8,
    #     "sentence_silence": 0.00,
    # }  argüman olarak alacak ve default değerler alacak , 

# save metodu olacak bir argüman alıyor , çıkış (ses) dosyasını kaydedecek ve ses dosyanın ismi unique olacak. uniqe ismi döndürecek.
# play_sound adında bir metot olacak -> txt dosyasını alacak , txt dosyasını ve dosyayı ses dosyayıla oynatacak şekilde yazılacak.
# elif fonksiyon oluşturacak create_unique()  ses dosyası için amacı uniqe isim oluşturacak ,  uuid veya datetime kullanılarak yapılacak.
```
